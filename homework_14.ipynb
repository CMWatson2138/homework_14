```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv(r'C:\Users\watso\Documents\Data_Science_Bootcamp\homework_13\bottle.csv')
pd.options.display.max_columns=74
df
```

## Removing redundant/unnecessary columns

```python
bottle_df = df.copy()
bottle_df = bottle_df.drop(['Cst_Cnt', 'Btl_Cnt', 'Sta_ID', 'Depth_ID', 'O2ml_L', 'STheta', 'O2Sat', 'BtlNum', 'RecInd', 'T_prec', 'T_qual', 'S_prec', 'S_qual', 'P_qual', 'O_qual', 'SThtaq', 'O2Satq', 'Chlqua', 'Phaqua', 'PO4q', 'SiO3qu', 'NO2q', 'NO3q', 'NH3q', 'C14A1p', 'C14A1q', 'C14A2p', 'C14A2q', 'DarkAp', 'DarkAq', 'MeanAs', 'MeanAp', 'MeanAq', 'IncTim', 'LightP', 'R_Depth', 'R_TEMP', 'R_POTEMP', 'R_SALINITY', 'R_SIGMA', 'R_SVA', 'R_DYNHT', 'R_O2', 'R_O2Sat', 'R_SIO3', 'R_PO4', 'R_NO3', 'R_NO2', 'R_NH4', 'R_CHLA', 'R_PHAEO', 'R_PRES', 'R_SAMP', 'DIC Quality Comment'], axis=1)
```

## Removing columns with excessive number of null values

```python
bottle_df = bottle_df.drop(['NH3uM', 'C14As1', 'C14As2', 'DarkAs', 'DIC1', 'DIC2', 'TA1', 'TA2', 'pH2', 'pH1'], axis=1)
bottle_df
```

## Removing Outliers

```python
mod_bot_df=bottle_df.copy()
# IQR
Q1 = mod_bot_df.quantile(0.25)
Q3 = mod_bot_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
mod_bot_df_out = mod_bot_df[~((mod_bot_df < (Q1 - 1.5 * IQR)) |(mod_bot_df > (Q3 + 1.5 * IQR))).any(axis=1)]
mod_bot_df_out.shape
```

## Standardized columns

```python
mod_bot_df_out.var()
```

```python
mod_bot_df_out['Depthm_log'] = np.log(mod_bot_df_out['Depthm'])
mod_bot_df_out['SiO3uM_log'] = np.log(mod_bot_df_out['SiO3uM'])
mod_bot_df_out['T_degC_log'] = np.log(mod_bot_df_out['T_degC'])
mod_bot_df_out['Oxy_Âµmol/Kg_log'] = np.log(mod_bot_df_out['Oxy_Âµmol/Kg'])
mod_bot_df_out['NO3uM_log'] = np.log(mod_bot_df_out['NO3uM'])
mod_bot_df_out=mod_bot_df_out.drop(['Depthm', 'SiO3uM', 'T_degC', 'NO3uM', 'Oxy_Âµmol/Kg'], axis=1)
mod_bot_df_out
```

## Replaced NaN values with mean

```python
mod_bot_df_out.fillna(mod_bot_df_out.mean(), inplace=True)
mod_bot_df_out
```

## 6.	What are the other ways to assess the performance of the model besides accuracy? What approach makes the most sense with the diabetes dataset? Why? 

As discussed in class it is probably more important to over-identify people as having diabetes when they don't, versus being most accurate or under-identifying them. ## 7.	Is the KNN model or the multiple logistic regression model more performant for the diabetes dataset? How can you tell? 

The KNN model is more performant for the diabetes dataset than the multiple logistic regression model.  I can tell this because recall and accuracy numbers are both superior for this model.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']

#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=45)

#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train,y_train)
y_predicted = knn.predict(X_test)
print(y_predicted)
```

```python
print(knn.score(X_test, y_test))
```

```python
confus_matrix=confusion_matrix(y_test, y_predicted)
confus_matrix
```

```python
plot_confusion_matrix(knn, X_test, y_test)
```

```python
print(classification_report(y_test, y_predicted))
```
